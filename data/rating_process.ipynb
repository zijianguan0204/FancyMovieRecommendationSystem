```python
import pandas as pd
import json

df = pd.read_csv('ratings.csv')
# df.drop(columns=['timestamp'],inplace=True)
# df = df.reset_index(drop=True).set_index(['userId','movieId'])
df.head(5)
print(len(df))
```

```python
print(len(df))
df.drop_duplicates(subset=['userId','movieId'],inplace=True)
print(len(df))
df.dropna(axis=0,subset=['rating'])
print(len(df))
df.head(50)
```

```python
df['rating'] = df['rating'].astype(int)
df['rating'] = df['rating'].map(lambda x: max(x,1))
print(df['rating'].unique())
```

```python
df = df.reset_index(drop=True).set_index(['userId','movieId'])

df.to_csv('rating_processed.csv')
```
